```python
### PROCESS PATIENTS RECORDS FOR CONDENSING

# Load Patient Records
rcd_full = pd.read_csv(os.path.join(path, table))
# Condensed Patient Records
record = pd.DataFrame(columns = ['SubjID', 'Timestamp'])

# Prevent Skipping Indices when Consolidating Grouped Timepoints
i = 0

for entry in range(len(rcd_full)):

# Check if Entry Times are Grouped
    # Add First Row Data of Reports and of Each New Patient
        # Data of Each Patient is Grouped
    if entry == 0 or rcd_full.loc[entry, 'SubjID'] != rcd_full.loc[entry-1, 'SubjID']:
        record.loc[i] = rcd_full.loc[entry,'SubjID'], pd.Timestamp(rcd_full.loc[entry,'Timestamp'])
    # Grouped Times Considered Within 5 min of the Previous
    elif pd.Timestamp(rcd_full.loc[entry, 'Timestamp']) <= pd.Timestamp(rcd_full.loc[entry-1,'Timestamp']) + pd.Timedelta('5 min'):
        continue
    # If not Grouped, Add Subject and Timepoint to New DataFrame
    else:
        record.loc[i] = rcd_full.loc[entry,'SubjID'], pd.Timestamp(rcd_full.loc[entry,'Timestamp'])
    
    i += 1

# Save Consolidated DataFrame as CSV
record.to_csv(os.path.join(dest, dest_ext, file_name), index = False)
```
